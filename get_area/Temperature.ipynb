```python
all_area = pd.read_csv("./AllArea.csv")
exe_path = "./../static/chromedriver_mac"
xp = '//*[@id="maincontent"]/div/div[3]/div[2]/div[2]/div[1]/div[3]/table/tbody'
city_list = all_area['city']
null_data = [""]*12


def get_temp(city, xp):  
#     url = "https://www.holiday-weather.com/" + "_".join(city.lower().split(" ")) + "/averages"
    url = "https://www.timeanddate.com/weather/" + country + "/" + city +"/climate"
    driver.get(url)
    time.sleep(2)
    try:
        data = driver.find_element_by_xpath(xp).text
        data = data.split("\n")[1].split(" ")[-12:]
    except:
        return []
    return data

def init_temp():
    temp_data = dict()
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    temp_data['city'] = list()
    for m in months: temp_data[m] = list()
    return temp_data

```

```python
success = 0
temp_data = init_temp()
error_list = list()
for index, city in enumerate(city_list):
    temp = get_temp(city, xp)
    if (len(temp) != 12): 
        print(success, index+1, "Error at: ", temp, len(temp), city)
        error_list.append(city)
        temp = null_data
    else: 
        success += 1
        print(success, index+1, "SUCCESS AT: ", city)
    temp_data['city'].append(city)
    for ind, mon in enumerate(months):
        temp_data[mon].append(temp[ind])
#    if index > 30: break
```
