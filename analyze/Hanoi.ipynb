```python
import pandas as pd
import numpy as np
```

```python
from datetime import datetime


# from datetime import datetime

# datetime_str = '09/19/18 13:55:26'


# # printed in default format
def convert_timestamp(a=''):
    datetime_object = datetime.strptime(a, '%Y-%m-%d_%H')
    return datetime_object
# a = '2021-12-05_12'
# convert_timestamp(a)
```

```python
df = pd.read_csv('../ProcessedAQI/Final/aqi_all.csv')
df['timestamp'] = df['timestamp'].map(lambda x: convert_timestamp(x))
df1 = df.loc[df['city'] == 'Hanoi']
df1
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20,10))
x = df1['timestamp'].tolist()
y = df1['AQI'].tolist()
plt.yticks(np.arange(0, max(y), 10))
plt.title('Hanoi AQI form 05/12/2021 to 12/12/2021')
plt.ylabel('AQI')
plt.xlabel('timestamp')
plt.plot(x,y)
plt.show()
```

```python
df2 = df.loc[df['city'] == 'Stockholm']
fig = plt.figure(figsize = (20,10))
x = df2['timestamp'].tolist()
y = df2['AQI'].tolist()
plt.yticks(np.arange(0, max(y), 10))
# plt.yticks(np.arange(0, max(y), 10))
plt.title('Stockholm AQI form 05/12/2021 to 12/12/2021')
plt.ylabel('AQI')
plt.xlabel('Timestamp')
plt.plot(x,y)
plt.show()
```
